```python
#IMPORTING ALL THE NECESSARY LIBRRARIES
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('BTC.csv')

# Convert 'Date' to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Focusing on 'Close' column
closeprices = df['Close']

print("Total number of records:",len(closeprices))
print("First 5 Close prices:\n", closeprices.head())
```

```python
# Mean
meanclose=closeprices.mean()
print("Mean (Average) of Close prices:",meanclose)

# Median
medianclose=closeprices.median()
print("Median of Close prices:", medianclose)

# Mode
modeclose=closeprices.mode()[0]
print("Mode of Close prices:", modeclose)
```

```python
# Range
rangeclose=closeprices.max()-closeprices.min()
print("Range of Close prices:", rangeclose)

# Variance
varianceclose=closeprices.var()
print("Variance of Close prices:", varianceclose)

# Standard Deviation
stdclose=closeprices.std()
print("Standard Deviation of Close prices:",stdclose)

# Coefficient of Variation
cvclose=(stdclose/meanclose)*100
print("Coefficient of Variation (CV):", cvclose,"%")
```

```python
plt.figure(figsize=(10, 5))
plt.hist(closeprices,bins=40,color='lightgreen',edgecolor='purple')
plt.title('Distribution of Bitcoin Close Prices')
plt.xlabel('Close Price (USD)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
#summary of results till now
print("SUMMARY TILL NOW ")
print("MEAN:",meanclose)
print("MEDIAN:",medianclose)
print("MODE:",modeclose)
print("RANGE:",rangeclose)
print("STD DEV SIGMA:",stdclose)
print("VARIANCE SIGMA SQUARE:",varianceclose)
print("Coeff of Variation:",cvclose)
```

```python
import seaborn as sns#USING SEABORN FOR BETTTR GRAPHICS
import matplotlib.pyplot as plt
# Scatter plot
plt.figure(figsize=(8,6))
sns.scatterplot(data=df,x='Open',y='Close')
plt.title('SCATTER PLOT FOR O AND C PRICE S')
plt.xlabel('PRICE OPEN')
plt.ylabel('PRICE CLOSE')
plt.grid(True)
plt.show()

```

```python
#LINEAR REGRESS
#CREATINGF REGRESSION LINE PLOT
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt

# X and Y values
X=df[['Open']]
Y=df['Close']
xtr,xts,ytr,yrs=train_test_split(X,Y,test_size=0.1,random_state=42)



# MODEL C AND T
model = LinearRegression()
model.fit(xtr, ytr)
eval = model.predict(xts)
r2_test = r2_score(yrs,eval)
print(f"R2 Score:{r2_test}")

#PLOTTINGG
plt.plot(df['Open'], model.predict(X), color='blue')
plt.xlabel('Open')
plt.ylabel('Close')
plt.title('Regression Line: Open vs Close')
plt.show()
```

```python
from scipy.stats import ttest_ind
# Convert D to DT
df['Date'] = pd.to_datetime(df['Date'])

# Filter year wisse
btc2021=df[df['Date'].dt.year==2021]['Close']
btc2022=df[df['Date'].dt.year==2022]['Close']
# T-test
t_stat=ttest_ind(btc2021,btc2022,equal_var=False)

# RESULTINT VALS
print("Independent t-test between 2021 and 2022 BTC Close Prices:")
print("t-VALUE UPTO 4TH DECIMAL=%.4f"%t_stat.statistic)
mean2021=btc2021.mean()
mean2022=btc2022.mean()
```
